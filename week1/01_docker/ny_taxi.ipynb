```python
hw = pd.read_csv("green_tripdata_2019-01.csv",nrows=100)

hw.head()
```

```python
table_name = "green_tripdata_trip"
hw_iter = pd.read_csv('green_tripdata_2019-01.csv', iterator=True, chunksize=100000)

hw = next(hw_iter)

hw.lpep_pickup_datetime = pd.to_datetime(hw.lpep_pickup_datetime)
hw.lpep_dropoff_datetime = pd.to_datetime(hw.lpep_dropoff_datetime)

hw.head(n=0).to_sql(name=table_name, con=db, if_exists='replace')

hw.to_sql(name=table_name, con=db, if_exists='append')

while True: 

        try:
            t_start = time()
            
            hw = next(hw_iter)

            hw.lpep_pickup_datetime = pd.to_datetime(hw.lpep_pickup_datetime)
            hw.lpep_dropoff_datetime = pd.to_datetime(hw.lpep_dropoff_datetime)

            hw.to_sql(name=table_name, con=db, if_exists='append')

            t_end = time()

            print('inserted another chunk, took %.3f second' % (t_end - t_start))

        except StopIteration:
            print("Finished ingesting data into the postgres database")
            break
```

```python
hw3 = """SELECT
    CAST(lpep_pickup_datetime AS DATE) AS "puday",
	CAST(lpep_dropoff_datetime AS DATE) AS "doday",
	COUNT(1)
FROM
    green_trips t
WHERE CAST(lpep_pickup_datetime AS DATE)='2019-01-15' AND CAST(lpep_dropoff_datetime AS DATE)='2019-01-15'
GROUP BY "puday","doday"

LIMIT 100;"""

pd.read_sql(hw3, con=db)
```

```python
hw4 = """SELECT 
CAST(lpep_pickup_datetime AS DATE),
CAST(lpep_dropoff_datetime AS DATE),
trip_distance 
FROM green_trips 
WHERE trip_distance = ( SELECT MAX(trip_distance) FROM green_trips )"""

pd.read_sql(hw4, con=db)
```

```python
hw5 = """SELECT
    CAST(lpep_pickup_datetime AS DATE) AS "puday",
	COUNT(passenger_count) AS "passenger_count_2"
FROM
    green_trips t
WHERE "passenger_count" = 2 AND CAST(lpep_pickup_datetime AS DATE)='2019-01-01'

GROUP BY "puday",passenger_count

LIMIT 100;"""
pd.read_sql(hw5, con=db)
```

```python
hw5_2 = """SELECT
    CAST(lpep_pickup_datetime AS DATE) AS "puday",
	COUNT(passenger_count) AS "passenger_count_3"
FROM
    green_trips t
WHERE "passenger_count" = 3 AND CAST(lpep_pickup_datetime AS DATE)='2019-01-01'

GROUP BY "puday",passenger_count

LIMIT 100;"""

pd.read_sql(hw5_2, con=db)
```

```python
hw6 = """
SELECT
    MAX(tip_amount),
    zpu."Zone" AS "pickup_loc",
    CONCAT(zdo."Borough", '/', zdo."Zone") AS "dropoff_loc"
FROM
    green_trips t LEFT JOIN trip zpu
        ON t."PULocationID" = zpu."LocationID"
    LEFT JOIN trip zdo
        ON t."DOLocationID" = zdo."LocationID"
WHERE zpu."Zone" = 'Astoria' 
GROUP BY tip_amount,"pickup_loc","dropoff_loc"

ORDER BY tip_amount DESC
LIMIT 1;"""

pd.read_sql(hw6, con=db)
```
